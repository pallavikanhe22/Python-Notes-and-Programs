# **Dictionary in Python**
* A dictionary is a collection of key : value pairs
* Does not allow duplicate keys
* Values can be of any datatypes



* Type - Mapping
* Syntax - {Key : Value}
* Mutable - Yes
* Ordered - Yes
* Duplicate Keys - No
* Values allowed - Any data type


Example :

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
print(student)
```

Using dict() Constructor :



```python
my_dict = dict(name="pallavi", age=21, marks=85)
print(my_dict)
```

# **Functions**

1. Add new key - value

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
student ["marks"] = 92
print(student)
```

2. update () - Update multiple items

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
student.update({"city":"mumbai", "age":25})
print(student)
```

3. pop () - Removing items


```python
student = {"name":"pallavi", "age":21, "city":"pune"}
(student.pop("age"))
print(student)
```

4. popitem - remove last inserted item

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
student.popitem()
print(student)
```

# **Dictionary Methods**

1. keys () - Returns all keys

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
print(student.keys())
```

2. values () - Returns all values

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
print(student.values())
```

3. items () - Returns list of pairs (key, values)

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
print(student.items())
```

5. get () - Safe key access

* Using get() - safe access (returns None or default if key missing)
* Raises keyerror if key not found present)

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
print(student.get("name"))
print(student.get("gender"))
print(student.get("NA"))
```

# **Looping Through Dictionary**

1. Loop over keys

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
for key in student:
  print(key)
```

2. Loop Over Values

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
for val in student.values():
  print(val)
```

3. Loop Over Key-Value Pairs

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
for k, v in student.items():
  print(k, ":", v)
```

# **Check if a key exists**

```python
student = {"name":"pallavi", "age":21, "city":"pune"}
print("name" in student)
print("marks" in student)
```

# **Dictionary with different datatypes**

```python
employee = {"name":"pallavi", "skills": ["python", "SQL"], "details": {"dept":"IT", "year":2025}}
print(employee["skills"][1])
print(employee["details"] ["dept"])
```

# **Dictionary from two lists (Using ZIP ())**

```python
names = ["pallavi", "vishal", "ram", "pooja"]
marks = [89, 90, 87, 69]
result = dict(zip(names, marks))
print(result)
```
