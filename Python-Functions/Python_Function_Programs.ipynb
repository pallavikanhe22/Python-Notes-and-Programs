####3. Write a function to find the factorial of a number.

Description: Implement a function that computes the factorial of a given number using recursion or iteration.

```python
def factorial(n):
  result = 1
  for i in range(1, n + 1):
    result = result * i
  return result
```

```python
value = 4
print("Factorial of", value, "is =", factorial(value))
```

####4. Design a function to check if a given number is prime.

Description: A prime number is greater than 1 and has no divisors other than 1 and itself.

```python
def prime(num):
    if num <= 1:
        return "Not Prime Number"

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return "Not Prime Number"

    return "Prime Number"
```

```python
print(prime(11))
```

####6. Write a function to compute the nth Fibonacci number.

Description: Fibonacci sequence: 0, 1, 1, 2, 3, 5, 8...

```python
def fibonacci(n):
  a, b = 0, 1
  for i in range(n):
    a, b = b, a + b
  return a
```

```python
print(fibonacci(4))
```

####8. Design a function to count the number of vowels in a string.

Description: Count a, e, i, o, u (both lowercase & uppercase).

```python
def count_vowels(text):
  vowels = "aeiouAEIOU"
  count = 0
  for ch in text:
    if ch in vowels:
      count += 1
  return count
```

```python
print(count_vowels("Hello World"))
```

####9. Write a function to calculate the factorial of each number in a given list.

Description: Return a new list containing factorial of each number.

```python
def factorial_list(numbers):
  def fact(n):
    result = 1
    for i in range(1, n + 1):
      result *= i
    return result
  return [fact(num) for num in numbers]
```

```python
print(factorial_list([3, 4, 5]))
```

####15. Create a function to generate a list of prime numbers within a specified range.

Description: Takes start & end, returns list of primes.

```python
def prime_range(start, end):
  primes = []
  for num in range(start, end + 1):
    if num > 1:
      for i in range(2, int(num ** 0.5) + 1):
          if num % i == 0:
             break
      else:
          primes.append(num)
  return primes
```

```python
print(prime_range(10, 30))
```

####17. Write a function to remove duplicate elements from a list.

Description: Preserve original order.

```python
def remove_duplicates(list):
  unique_list = []
  for item in list:
      if item not in unique_list:
        unique_list.append(item)
  return unique_list
```

```python
print(remove_duplicates([1, 2, 3, 2, 3, 6, 7, 8]))
```

####18. Create a function to compute the sum of squares of all numbers from 1 to n.

Description: sum = 1² + 2² + ... + n²

```python
def sum_of_squares(n):
  total = 0
  for i in range(1, n + 1):
    total += i * i
  return total
```

```python
print(sum_of_squares(5))
```

####19. Design a function to calculate the distance between two points in a 2D plane.

Description: Use formula → distance = √((x2-x1)² + (y2-y1)²)

```python
import math

def distance(x1, y1, x2, y2):
  return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
```

```python
print(distance(0, 0, 3, 4))
```

####20. Write a function to find the mode (most frequent element) in a list.

Description: If multiple modes exist, return any one.

```python
def find_mode(list):
  frequency = {}
  for item in list:
      frequency[item] = frequency.get(item, 0) + 1
  return max(frequency, key=frequency.get)
```

```python
print(find_mode([2, 3, 2, 5, 3, 2]))
```
