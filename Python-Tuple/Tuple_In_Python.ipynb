# **Tuple in Python**

* A tuple is an ordered collection of elements
* Immutable - once created, you cannot change, add, or remove element
* Uses round brackets ()
* Can store different data types (int, float, string, list, tuple, etc)


Example :

```python
my_tuple = (1, "Apple", 3.5, [1, 2, 3], {"a", "b"})
print(my_tuple)
```

# **Indexing**

* Accessing tuple elements
* Index start from 0
* Negative index start from -1

Example :

```python
t = ('a', 'b', 'c', 'd')
print(t[0])
print(t[2])
print(t[-1])
print(t[-3])
```

```python
t = (1, (2, 3), (4, 5, 6))
print(t[2] [1])
```

# **Slicing**

Used to extract a part of the tuple

Example :

```python
t = (10, 20, 30, 40, 50)
print(t[1:4])
print(t[:3])
print(t[2:])
print(t[:])
print(t[-3:])
print(t[::-1])
```

# **Tuple Method**

Tuples have two important methods
1. index ()
2. count ()


###1. index ()

Returns first index of a given value

```python
t = (1, 2, 3, 4, 6, 5)
print(t.index(2))
```

###2. count ()

Returns numbers of occurrences of a value

```python
t = (1, 2, 2, 2, 4)
print(t.count(2))
```

# **Tuple Functions**

1. len () - Length of tuple

```python
t = (1, 2, 4, 2, 6, 3, 8)
print(len(t))
```

2. max () - Largest value

```python
t = (1, 2, 4, 2, 6, 3, 8)
print(max(t))
```

3. min () - Smallest value

```python
t = (1, 2, 4, 2, 6, 3, 8)
print(min(t))
```

4. sorted () - Returns sorted list

```python
t = (1, 2, 4, 2, 6, 3, 8)
print(sorted(t))
```

5. sum () - Total of numbers

```python
t = (1, 2, 4, 2, 6, 3, 8)
print(sum(t))
```

6. average () - Custom calculation

```python
t = (1, 2, 4, 2, 6, 3, 8)
avg = sum(t) / len(t)
print(avg)
```

# **Using Tuple with ZIP()**

```python
cities = ["Delhi", "Patna", "Cuttack"]
rivers = ["Yamuna", "Ganga", "Mahanandi"]

result = list(zip(cities,rivers))
print(result)
```

# **When to Use Tuple**

1. Data must not change
2. You want faster performance
3. You want to return multiple values from function
4. You want to use tuple as a dictionary key
