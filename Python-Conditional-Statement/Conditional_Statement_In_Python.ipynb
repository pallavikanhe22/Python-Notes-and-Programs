# **Conditional Statements in Python**

* Conditional statements are used to control the flow of a program.
* They allow the program to execute specific blocks of code based on certain conditions.
* They help the program make decisions depending on different situations.

Types of Conditional Statements
1. if
2. if-else
3. if-elif-else
4. nested if


# **1. if Statement**

* The if statement is used to execute a block of code only when a given condition is true.
* If the condition is false, Python simply skips that block.


Syntax :


```python
if condition:
    # code to execute when condition is true
```

Example :

```python
x = 10
if x > 5:
  print("x is greater than 5")
```

# **2. if-else Statement**

* The if-else statement is used to execute one block of code when the condition is true, and execute the else block when the condition is false.
* It ensures that one of the two blocks will always run.

Syntax :


```python
if condition:
    # code if true
else:
    # code if false
```

Example :

```python
age = 18
if age >= 18:
  print("Eligible to vote")
else:
  print("Not eligible to vote")
```

# **3. if-elif-else statement**

* The if-elif-else statement is used when you need to check multiple conditions one after another.
* Python evaluates each condition in order:

* If a condition becomes true, its block is executed and the chain stops.

* If none of the conditions are true, the else block runs (if provided).

* This structure is used when only one condition should be selected from many possibilities.

Syntax :


```python
if condition1:
    # code if condition1 is true
elif condition2:
    # code if condition2 is true
else:
    # code if all conditions are false
```

Example :

```python
marks = int(input("Enter a marks : "))
if marks >= 90:
  print("Grade A")
elif marks >= 75:
  print("Grade B")
elif marks >= 50:
  print("Grade C")
else:
  print("Fail")
```

# **4. Nested if Statement**

* A nested if means placing an if statement inside another if statement.
It is used when one condition depends on another condition.
* This structure is useful for checking step-by-step, multi-level, or hierarchical conditions

Syntax :

```python
if condition1:
    # code for condition1

    if condition2:
        # code for condition2 inside condition1

        if condition3:
            # code for condition3 inside condition2
    else:
        # code when condition2 is false
else:
    # code when condition1 is false
```

Example :

```python
num = 25
if num > 0:
  print("Positive number")
  if num % 2 == 0:
    print("Even number")
  else:
    print("Odd number")
else:
  print("Negative number")
```
