Write a Python program that takes an integer as input and checks the following conditions using nested if:

* If it is even, check if it is greater than 50.
* If it is odd, check if it is a multiple of 5.
* If the number is negative, print "Number is negative."



```python
num = int(input("Enter a number : "))

if num > 0:
  print("Number is positive")
  if num % 2 == 0:
    print("Number is even")

    if num > 50:
      print("Number is greater than 50")
    else:
      print("Number is 50 or less")
  else:
    print("Number is odd")

    if num % 5 == 0:
      print("Number is a multiple of 5")
    else:
      print("Number is not a multiple of 5")

elif num < 0:
  print("Number is negative")
else:
  print("Number is zero")
```

```python
a = int(input("Enter side 1 : "))
b = int(input("Enter side 2 : "))
c = int(input("Enter side 3 : "))

# check validity first
if a + b > c and a + c > b and b + c > a:

    if a == b == c:
        print("Equilateral Triangle")
    elif a == b or b == c or a == c:
        print("Isosceles Triangle")
    else:
        print("Scalene Triangle")

else:
    print("Not a valid triangle")
```

####13. Password Strength Checker

Write a Python program that takes a password as input and checks its strength:
* If the length is less than 6 characters, print "Weak password."
* If the length is 6 or more, check:
* If it contains numbers and special characters, print "Strong password."
* Else, print "Moderate password, add special characters."

```python
password = input("Enter password: ")

if len(password) < 6:
    print("Weak password.")

else:
    has_number = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_number = True
        if not ch.isalnum():   # special character
            has_special = True

    if has_number and has_special:
        print("Strong password.")
    else:
        print("Moderate password, add special characters.")
```
