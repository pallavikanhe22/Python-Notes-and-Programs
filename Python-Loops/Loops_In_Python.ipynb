# **Loops in Python**

* Loops in Python are used to repeat a block of code multiple times
* They help automate tasks that need to run again and again without writing the same code repeatedly
* Loops continue executing until a specific condition becomes false

Types of Loops :

1. for Loop
2. while Loop

# **1. for Loop**

* The for loop in python is used to iterate over a sequence (like list, tuple, string, or range)
* It automatically takes each item from the sequence one by one and executes the loop block for each element


Key Point :

A for loop is used when the number of iterations is known or when you are looping through a collection

Syntax :

```python
for variable in sequence:
    # code runs for each item
```

Example 1 :

```python
# Print numbers from 1 to 5

for i in range (1, 6):
  print(i)
```

Example 2 :

```python
# Print each element in a list

numbers = [10, 20, 30, 40]
for n in numbers:
  print(n)
```

Example 3 :

```python
# Print characters of a string

for ch in "python":
  print(ch)
```

Example 4 :

```python
# Find sum of list elements

total = 0
nums = [4, 6, 8, 2]
for n in nums:
  total += n
  print("Sum = ", total)
```

# **2. while Loop**

* The while loop executes a block of code as long as the given condition remain true
* It repeatedly checks the condition before each iteration
* As soon as the condition becomes false, the loop stops

Key Point :

A while loop is used when the number of iterations is not fixed and depends on the condition

Syntax :

```python
while condition:
    # code runs until condition becomes false
```

Example 1 :

```python
# Print numbers from 1 to 5

i = 1
while i <= 5:
  print(i)
  i += 1
```

Example 2 :

```python
# Countdown

n = 5
while n > 0:
  print(n)
  n -= 1
```

# **Loop Control Statements**



##1. break

The break statement is used to immediately stop or exit the loop, even if the loop condition is still true.
Once break executes, control moves out of the loop.

Use : When you want to terminate the loop early.

Example :

```python
for i in range(1, 10):
    if i == 4:
        break
    print(i)
```

##2. continue

The continue statement skips the current iteration of the loop and moves directly to the next iteration.
It does not stop the loop — it only skips one cycle.

Use : When you want to ignore specific values/conditions but keep looping.

Example :

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

##3. pass

The pass statement does nothing.
It is a placeholder used when a statement is required syntactically but you don’t want to perform any action.

Use : For creating empty loops, empty functions, or for future code.

Example :

```python
for i in range(1, 6):
    if i == 3:
        pass
```
