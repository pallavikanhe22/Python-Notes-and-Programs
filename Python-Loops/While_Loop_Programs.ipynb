####1. Write a Python program to print the sum of all even numbers between 1 and 100 using a while loop.

* Output : Sum of even numbers = 2550

```python
i = 1
total = 0

while i <= 100:
  if i % 2 == 0:
    total += i
  i += 1

print("Sum of even number = ", total)
```

####2. Write a Python program to reverse a number using a while loop.

Example :
* Input : 1234
* Output : 4321

```python
num = int(input("Enter a number : "))
rev = 0

while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10

print("Reversed number : ", rev)
```

####3. Write a Python program to find the sum of digits of a given number using a while loop.

Example :
* Input : 567
* Output : 18 (5 + 6 + 7 = 18)

```python
num = int(input("Enter a number : "))
total = 0

while num > 0:
  digit = num % 10
  total += digit
  num //= 10

print("Sum of digits : ", total)
```

####4. Write a Python program to find the greatest common divisor (GCD) of two numbers using a while loop.

Example :
* Input : 36, 60
* Output : 12

```python
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))

while b != 0:
  a, b = b, a % b

print("GCD =", a)
```

####5. Write a Python program to calculate the power of a number using a while loop (without using the ** operator).

Example :
* Input : base = 2, exponent = 4
* Output : 16 (because 2^4 = 16)

```python
base = int(input("Enter base : "))
exp = int(input("Enter exponent : "))

result = 1
i = 1

while i <= exp:
  result *= base
  i += 1

print("Result =", result)
```
