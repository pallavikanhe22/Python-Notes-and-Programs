# **Nested Slicing Questions**

####11. Write a program to access the second element from the first nested list.
Input : [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

Output : 2


```python
lists = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(lists[0][1])
```

####12.  Write a program to extract the last element from the last nested list.
Input : [['a', 'b', 'c'], ['d', 'e', 'f'], ['g', 'h', 'i']]

Output : i


```python
lists = [['a', 'b', 'c'], ['d', 'e', 'f'], ['g', 'h', 'i']]
print(lists[-1][-1])
```

####13.  Write a program to extract the second and third lists from a given nested list.
Input : [[10, 20], [30, 40], [50, 60], [70, 80]]

Output : [[30, 40], [50, 60]]


```python
lists = [[10, 20], [30, 40], [50, 60], [70, 80]]
print(lists[1:3])
```

####14. Write a program to extract the second element from all nested lists.
Input : [[1, 9], [2, 8], [3, 7], [4, 6]]

Output : [9, 8, 7, 6]


```python
lists = [[1, 9], [2, 8], [3, 7], [4, 6]]
result = [x[1] for x in lists]
print(result)
```

####15. Write a program to reverse all sublists in a nested list.
Input : [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

Output : [[3, 2, 1], [6, 5, 4], [9, 8, 7]]


```python
lists = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
result = [sub[::-1] for sub in lists]
print(result)
```
